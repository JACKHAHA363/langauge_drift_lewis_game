```python
# Show the vocab change and shift during gumbel
```

```python
import torch
vocab_change = torch.load('../../vocab_change.pth')
log_freq = 50
monitor_word = 22
target_word = [20,21,22,23]
```

```python
speak_conf_mats = vocab_change['speak'].numpy()
steps = [(i+1)*log_freq for i in range(len(speak_conf_mats))]
print("number of steps", len(steps))
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(7,5))
for target in target_word:
    target_probs = [conf_mat[monitor_word-1, target-1] for conf_mat in speak_conf_mats]
    line, = ax.plot(steps, target_probs)
    line.set_label('word {}'.format(target))
ax.legend(fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
ax.set_xlabel('Steps', fontsize=15)
ax.set_ylabel('Prob', fontsize=15)
```
